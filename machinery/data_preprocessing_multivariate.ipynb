```python
from covid19 import get_death_data
from covid19 import get_population_data
from covid19 import get_first_dt
```

```python
import pandas as pd
import numpy as np
```

```python
data_deaths = get_death_data()
```

```python
get_first_dt(data_deaths, 't1', 1000).dropna()
```

```python
multivariate = pd.DataFrame([])
for i in np.arange(0, 1000, 1):
    nm = 't' + str(i)
    multivariate[nm] = get_first_dt(data_deaths, nm, i)[nm]
```

```python
multivariate
```

```python
def get_days(dt):
    try:
        days = dt.days
    except:
        days = None
    return days
```

```python
multivariate2 = pd.DataFrame([])
for i in np.arange(0, 995, 5):
    nm = 't' + str(i)
    nm1 = 't' + str(i+5)
    multivariate2[nm] = (multivariate[nm1] - multivariate[nm]).apply(get_days)
```

```python
multivariate2
```

```python
multivariate2.mean().plot()
```

```python
multivariate2.mean().hist(bins = 500)
```

```python
multivariate2.to_csv("waiting_times.csv")
```
